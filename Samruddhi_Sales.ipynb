```python
import pandas as pd

df = pd.read_excel("SaleReport_10_08_24_to_30_06_26.xlsx")
df.head()

df1 = pd.read_excel("Sale_Report_01-04-2021_to_08-08-2024.xls")
df1.head()
# Combine both files
df = pd.concat([df, df1], ignore_index=True)

# Remove duplicate rows
df = df.drop_duplicates()

print("Total Rows:", len(df))
```

```python

# Convert Date
df['Date'] = pd.to_datetime(
    df['Date'],
    dayfirst=True
)


# Create Year and Month columns
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
monthly = (
    df.groupby(['Year', 'Month', 'Item Name'], as_index=False)
      .agg(
          Quantity=('Quantity', 'sum'),
          Amount=('Amount', 'sum')
      )
)

print(monthly.shape)
monthly.to_excel("Monthly_Product_Sales.xlsx", index=False)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

monthly['ProductID'] = le.fit_transform(monthly['Item Name'])
```

```python
import joblib

joblib.dump(le, "product_encoder.pkl")
```

```python
X = monthly[
    [
        'ProductID',
        'Year',
        'Month'
    ]
]

y = monthly['Quantity']
```

```python
train = monthly[monthly['Year'] <= 2025]
test = monthly[monthly['Year'] == 2026]

X_train = train[X.columns]
y_train = train['Quantity']

X_test = test[X.columns]
y_test = test['Quantity']
```

```python
!pip install catboost
```

```python
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from catboost import CatBoostRegressor
```

```python
rf = RandomForestRegressor(
    n_estimators=300,
    random_state=42
)

rf.fit(X_train,y_train)
```

```python
xgb = XGBRegressor(
    n_estimators=300,
    learning_rate=0.05,
    max_depth=6,
    random_state=42
)

xgb.fit(X_train,y_train)
```

```python
cat = CatBoostRegressor(verbose=False)

cat.fit(X_train,y_train)
```

```python
from sklearn.metrics import mean_absolute_error
import numpy as np

models = {
    "RF": rf,
    "XGB": xgb,
    "CAT": cat
}

for name, model in models.items():

    pred = model.predict(X_test)

    mae = mean_absolute_error(y_test,pred)

    rmse = np.sqrt(((y_test-pred)**2).mean())

    print(name,mae,rmse)
```

```python
import joblib

joblib.dump(xgb,"inventory_model.pkl")
```

```python
import streamlit as st
import joblib

model = joblib.load("inventory_model.pkl")
encoder = joblib.load("product_encoder.pkl")

product = st.selectbox(
    "Product",
    encoder.classes_
)

year = st.number_input("Year",2026)

month = st.selectbox(
    "Month",
    list(range(1,13))
)

if st.button("Predict"):

    pid = encoder.transform([product])[0]

    X = [[
        pid,
        year,
        month
    ]]

    pred = model.predict(X)

    st.success(
        f"Required Inventory : {round(pred[0])}"
    )
```
